Logistic Regression class defined

```python
class LogisticRegression:

  def __init__(self, learning_rate, iterations=100, w=None, b=0) :
    self.iterations=iterations
    self.learning_rate=learning_rate
    self.b=b
    self.w=w

  def sigmoid(self,x):
    return 1/(1 + np.exp(-x))

  def cross_entropy_loss(self,target,actual):
    return (-target * np.log(actual) - (1 - target) * np.log(1 - actual)).mean() 

  def GetActualValue( self ):     
        return  self.X.dot( self.w) + self.b

  def fit(self,X,Y):
    self.X=X
    self.Y=Y   
    
    self.samples, self.features = X.shape  
    if self.w is None :      
        self.w = np.zeros( self.features) 

    for i in range(self.iterations):
       print('For iteration :')
       print(i)
       self.update_weights()

  def update_weights(self):

    y=self.GetActualValue()
    z=self.sigmoid(y)
    error=z-self.Y.T
    dw=(np.dot(self.X.T,error))/self.samples
    db=np.sum(error)/self.samples

    loss = self.cross_entropy_loss(self.Y.T,z)
    print(f'loss: {loss} \t')  
    print(f'w,b before update: {(self.w,self.b)} \t')

    self.w -= self.learning_rate*dw
    self.b -= self.learning_rate*db

    print(f'w,b after update: {(self.w,self.b)} \t')
        

  def predict( self, X ) : 
        y=np.dot(X,self.w) + self.b
        Z = self.sigmoid(y)    
        Y = np.where( Z > 0.5, 1, 0 )        
        return Y   
```

Normalization of features

```python
def FeatureNormalization(features_dataset):
    minmax_col=list()
    num_features=len(features_dataset[0])
    for i in range(num_features):
      values=[row[i] for row in features_dataset]
      min=np.min(values)
      max=np.max(values)
      minmax_col.append([min,max])
    for row in features_dataset:
      for i in range(len(row)):
        row[i]=(row[i]-minmax_col[i][0])/(minmax_col[i][1]-minmax_col[i][0])
```

Function for creating confusion matrix and returning Accuracy Precision Recall

```python
def CalAccPrecisionRecall(y_test,y_pred):
  TP=FP=TN=FN=0
  for i in range(len(y_test)):
    if y_pred[i]==y_test[i]:
      if y_test[i]==1:
         TP=TP+1
      else:
         TN=TN+1   
    else:
      if y_test[i]==1:
        FN=FN+1
      else :
        FP=FP+1
   
  return (TP+TN)/(TP+FP+TN+FN),TP/(TP+FP),TP/(TP+FN)     
```
